**Question No.1** Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?.
Examples
stutter("incredible") ➞ "in... in... incredible?"

stutter("enthusiastic") ➞ "en... en... enthusiastic?"

stutter("outstanding") ➞ "ou... ou... outstanding?"

Hint :-Assume all input is in lower case and at least two characters long.


```python
# **Ans**

def stutter(word):
    word1 = word[0:2]
    print('"',word1,'... ',word1,'... ',str(word),'?','"')
    
word = input('please input a string:')
stutter(word)
```

**Question No.2** Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.
Examples
radians_to_degrees(1) ➞ 57.3

radians_to_degrees(20) ➞ 1145.9

radians_to_degrees(50) ➞ 2864.8


```python
# **Ans**

def Convert_to(radian):
    pi = 3.14159
    
    degree = radian * (180/pi)
    return degree
 

radian = 30
print("degree =",(Convert_to(radian)))
```

**Question No.3** In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.
Given a non-negative integer num, implement a function that returns True if num is a Curzon number, or False otherwise.
Examples
is_curzon(5) ➞ True
 2 ** 5 + 1 = 33
 2 * 5 + 1 = 11
 33 is a multiple of 11

is_curzon(10) ➞ False
 2 ** 10 + 1 = 1025
 2 * 10 + 1 = 21
 1025 is not a multiple of 21

is_curzon(14) ➞ True
 2 ** 14 + 1 = 16385
 2 * 14 + 1 = 29
 16385 is a multiple of 29


```python
# **Ans**

def check_CurzonNumber(N):
 
    powerTerm, productTerm = 0, 0
 
    # Find 2^N + 1
    powerTerm = pow(2, N) + 1
 
    # Find 2*N + 1
    productTerm = 2 * N + 1
 
    # Check for divisibility
    if (powerTerm % productTerm == 0):
        print("Yes")
    else:
        print("No")
 

if __name__ == '__main__':
     
    N = 7
    check_CurzonNumber(N)
 
    N = 14
    check_CurzonNumber(N)
```

**Question No.4** Given the side length x find the area of a hexagon.

Examples
1. area_of_hexagon(1) ➞ 2.6

2. area_of_hexagon(2) ➞ 10.4

3. area_of_hexagon(3) ➞ 23.4


```python
# **Ans**


import math
 

def hexagonArea(x):
     
    return ((3 * math.sqrt(3) * (x * x)) / 2);
     
   
if __name__ == "__main__" :
 
    # length of a side.
    x = 4
 
    print("Area:","{0:.4f}" .format(hexagonArea(x)))
```
